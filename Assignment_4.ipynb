```
import numpy
from statistics import mean


directory = os.path.join('sample_data', 'california_housing_train.csv') # Output: sample_data/california_housing_train.csv

# print(directory)

# if os.path.isfile(directory):
#   print("File ditemukan")

text_file = open(directory, "r") # "r" = moud read

# print(text_file.read())

lines = text_file.readlines()

# print(lines)

# print("Jumlah data (original) ", len(lines))

# for line in lines[:3]: # Print 3 baris pertama
#   print(line)

  # line = line.strip() #fungsi nya menghapus spasi/enter di depan suatu text

  # split_columns = line.split(",")

  # print(split_columns)

#parsed_values = []
housing, room, bedroom, population, household, mincome, mhouse = [], [], [], [], [], [], []

for index, line in enumerate(lines):
  split_columns = line.split(",")

  if index == 0: #menskip judul
    continue
  housing.append(float(split_columns[2]))
  room.append(float(split_columns[3]))
  bedroom.append(float(split_columns[4]))
  population.append(float(split_columns[5]))
  household.append(float(split_columns[6]))
  mincome.append(float(split_columns[7]))
  mhouse.append(float(split_columns[8]))

# parsed_values.append([float(split_columns[0]), float(split_columns[1]), float(split_columns[2]), float(split_columns[3]), float(split_columns[4]), float(split_columns[5]), float(split_columns[6]), float(split_columns[7]), float(split_columns[8])])

# print("Jumlah data (parsed) ", len(parsed_values))

# print(parsed_values)

def fungsi(attribute):
  print("Nilai Mean adalah                  :", mean(attribute))
  print("Standard Deviasi adalah:           :", numpy.std(attribute))
  print("Nilai Minimum adalah               :", min(attribute))
  print("Nilai Maximum adalah               :", max(attribute))
  print("Nilai Q1 (Kuartil pertama) adalah  :", numpy.percentile(attribute, 25))
  print("Nilai Q2 (Kuartil kedua) adalah    :", numpy.percentile(attribute, 50))
  print("Nilai Q3 (Kuartil ketiga) adalah   :", numpy.percentile(attribute, 75))
  for n in attribute:
    if n == None:
      print("Terdapat nilai kosong/bernilai 0")
    else:
      print("Tidak terdapat nilai kosong/bernilai 0")
    break
  print("\n")
  return 

print("1. Housing Median Age")
print("=====================")
fungsi(housing)
print("2. Total Rooms")
print("==============")
fungsi(room)
print("3. Total Bedrooms")
print("=================")
fungsi(bedroom)
print("4. Population")
print("=============")
fungsi(population)
print("5. Households")
print("=============")
fungsi(household)
print("6. Median Income")
print("================")
fungsi(mincome)
print("7. Median House Value")
print("=====================")
fungsi(mhouse)
```
